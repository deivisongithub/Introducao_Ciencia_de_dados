## Projeto de ICD - Parte 3

```python
import pandas as pd
import seaborn as sns
DataBase = pd.read_csv("healthcare-dataset-stroke-data.csv")
stroke_1 = DataBase[DataBase["stroke"] == 1]
stroke_0 = DataBase[DataBase["stroke"] == 0]
```

## Com relação a idade

```python
Age = DataBase[["age"]]
Age_0 = stroke_0[["age"]]
Age_1 = stroke_1[["age"]]
```

### Indivíduos que sofreram AVC

```python
sns.displot(Age_1,x="age",binwidth = 10)
```

### Indivíduos que Não sofreram AVC

```python
sns.displot(Age_0,x="age",binwidth = 10)
```

***Nos gráficos acima temos as ocorrência de AVCs conforme a idade,onde fica nítido que com um aumento da idade temos mais casos de AVC,seguido por uma queda brusca,mostrando assim uma correlação entre entre idade e a chance de AVC.*** 

### Por faixa étaria

```python
Age_1_count = Age_1.value_counts()
Age_1_count_jovens = Age_1_count[Age_1_count.index < 19].sum()
Age_1_count_adulto = Age_1_count[Age_1_count.index < 59].sum() - Age_1_count_jovens
Age_1_count_idoso = Age_1_count[Age_1_count.index > 58].sum()
df_Age_1_count = pd.DataFrame([{"Jovens": Age_1_count_jovens, "Adultos": Age_1_count_adulto, "Idosos":Age_1_count_idoso}])
df_Age_1_count = df_Age_1_count.rename({0 :""}, axis='index')
df_Age_1_count.plot.bar (rot=0, title='Idade em relação ao AVC', figsize=(7,5), ylabel='Ocôrrencia' , xlabel='Faixa étaria')
Age_0_count = Age_0.value_counts()
Age_0_count_jovens = Age_0_count[Age_0_count.index < 19].sum()
Age_0_count_adulto = Age_0_count[Age_0_count.index < 59].sum() - Age_0_count_jovens
Age_0_count_idoso = Age_0_count[Age_0_count.index > 58].sum()
df_Age_0_count = pd.DataFrame([{"Jovens": Age_0_count_jovens, "Adultos": Age_0_count_adulto, "Idosos":Age_0_count_idoso}])
df_Age_0_count = df_Age_0_count.rename({0 :""}, axis='index')
df_Age_0_count.plot.bar (rot=0, title='Idade em relação ao AVC', figsize=(7,5), ylabel='Não ocôrrencia' , xlabel='Faixa étaria')
```

***Logo percebemos que os idosos são os mais propensos a sofrerem AVC***

### Analisando a dispersâo

```python
sns.scatterplot(data = DataBase,x = "stroke",y = "age")
```

## Com relacão ao Nivel de glicose 

```python
glicose = DataBase[["avg_glucose_level"]]
glicose_0 = stroke_0[["avg_glucose_level"]]
glicose_1 = stroke_1[["avg_glucose_level"]]
```

### Indivíduos que sofreram AVC

```python
sns.displot(glicose_1,x="avg_glucose_level",binwidth = 10)
```

### Indivíduos que Não sofreram AVC

```python
sns.displot(glicose_0,x="avg_glucose_level",binwidth = 10)
```

***Ao analisar comparativamente os gráficos acima, nota-se uma uma tendência de aumento nos números de AVC com relação ao nivel de glicose***

```python
glicose_1_count = glicose_1.value_counts()
glicose_1_count_baixa = glicose_1_count[glicose_1_count.index < 71].sum()
glicose_1_count_normal = glicose_1_count[glicose_1_count.index < 100].sum() - glicose_1_count_baixa
glicose_1_count_alta = glicose_1_count[glicose_1_count.index > 125].sum()
df_glicose_1_count = pd.DataFrame([{"Hipoglicemia": glicose_1_count_baixa, "Normal": glicose_1_count_normal, "Diabetes":glicose_1_count_alta}])
df_glicose_1_count = df_glicose_1_count.rename({0 :""}, axis='index')
df_glicose_1_count.plot.bar (rot=0, title='Idivíduos que sofreram AVC', figsize=(7,5), ylabel='ocôrrencia de AVC' , xlabel='Nivel de glicose')
glicose_0_count = glicose_0.value_counts()
glicose_0_count_baixa = glicose_0_count[glicose_0_count.index < 71].sum()
glicose_0_count_normal = glicose_0_count[glicose_0_count.index < 100].sum() - glicose_0_count_baixa
glicose_0_count_alta = glicose_0_count[glicose_0_count.index > 125].sum()
df_glicose_0_count = pd.DataFrame([{"Hipoglicemia": glicose_0_count_baixa, "Normal": glicose_0_count_normal, "Diabetes":glicose_0_count_alta}])
df_glicose_0_count = df_glicose_0_count.rename({0 :""}, axis='index')
df_glicose_0_count.plot.bar (rot=0, title='Indivíduos que Não sofreram AVC', figsize=(7,5), ylabel='Não ocôrrencia de AVC' , xlabel='Nivel de glicose')
```

***Se pode inferir,pelos gráficos acima, que indíviduos que sofrem de diabetes,tem uma forte tendência a também sofrer de AVC***
