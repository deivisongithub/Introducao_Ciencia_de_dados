```python
import pandas as pd
DataBase = pd.read_csv("healthcare-dataset-stroke-data.csv")
DataBase = DataBase.dropna()
Stroke_1 = DataBase[DataBase["stroke"] == 1]
Stroke_1
```

## Agrupamentos

```python
from sklearn.cluster import KMeans
import seaborn as sns
def group_KMeans(df,X,Y,n):
    Group = df[[X,Y]]
    Kmeans = KMeans(n_clusters = n)
    Kmeans.fit(Group)
    sns.scatterplot(data = Group,x = X ,y = Y,hue = Kmeans.labels_)
```

### Idade e Glicose

```python
group_KMeans(DataBase,"age","avg_glucose_level",2)
```

```python
group_KMeans(Stroke_1,"age","avg_glucose_level",2)
```

### Idade e Bmi

```python
group_KMeans(DataBase,"age","bmi",3)
```

```python
group_KMeans(Stroke_1,"age","bmi",3)
```

### Glicose e Bmi

```python
group_KMeans(DataBase,"avg_glucose_level","bmi",2)
```

```python
group_KMeans(Stroke_1,"avg_glucose_level","bmi",2)
```
