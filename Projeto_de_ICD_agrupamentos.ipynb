```python
import pandas as pd
DataBase = pd.read_csv("healthcare-dataset-stroke-data.csv")
DataBase = DataBase.dropna()
DataBase
```

```python
import seaborn as sns
sns.scatterplot(data = DataBase,x = "age",y = "avg_glucose_level")
```

```python
sns.scatterplot(data = DataBase,x = "age",y = "bmi")
```

```python
sns.scatterplot(data = DataBase,x = "bmi",y = "avg_glucose_level")
```

## Agrupamentos

### Idade e Glicose

```python
from sklearn.cluster import KMeans
Group_Age_glicose = DataBase[["age","avg_glucose_level"]]
Kmeans_Age_glicose = KMeans(n_clusters = 2)
Kmeans_Age_glicose.fit(Group_Age_glicose)
```

```python
sns.scatterplot(data = Group_Age_glicose,x = "age",y = "avg_glucose_level",hue = Kmeans_Age_glicose.labels_)
```

### Idade e Bmi

```python
Group_Age_Bmi = DataBase[["age","bmi"]]
Kmeans_Age_Bmi = KMeans(n_clusters = 3)
Kmeans_Age_Bmi.fit(Group_Age_Bmi)
```

```python
sns.scatterplot(data = Group_Age_Bmi,x = "age",y = "bmi",hue = Kmeans_Age_Bmi.labels_)
```

### Glicose e Bmi

```python
Group_glicose_Bmi = DataBase[["avg_glucose_level","bmi"]]
Kmeans_glicose_Bmi = KMeans(n_clusters = 2)
Kmeans_glicose_Bmi.fit(Group_glicose_Bmi)
```

```python
sns.scatterplot(data = Group_glicose_Bmi,x = "avg_glucose_level",y = "bmi",hue = Kmeans_glicose_Bmi.labels_)
```
