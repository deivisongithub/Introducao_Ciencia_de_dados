## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
# ADICIONE O SEU CÓDIGO AQUI
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.2f' % x)
tabela_filmes = pd.read_csv("movies.csv")
tabela_filmes["Lucro"] = tabela_filmes["Worldwide Gross"] - tabela_filmes["Production Budget"]
tabela_genero = tabela_filmes.groupby("Major Genre").mean()
tabela_genero.sort_values("Lucro",ascending = False)[["Worldwide Gross","Production Budget","Lucro"]]
```

***Pode-se notar,com base no DataFrame acima, que o gênero mais lucrativo é o de Aventura***

### Qual o diretor dirigiu mais filmes?

```python
# ADICIONE O SEU CÓDIGO AQUI
pd.DataFrame(tabela_filmes["Director"].value_counts()[0:10])
```

***Segundo o DataFrame acima,o diretor que dirigiu mais filmes foi Steven Spielberg***

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
# ADICIONE O SEU CÓDIGO AQUI
tabela_filmes["Vezes Maior"] = tabela_filmes["Worldwide Gross"] / tabela_filmes["Production Budget"]
tabela_filmes.sort_values(by="Vezes Maior", ascending=False)[["Title","Worldwide Gross","Production Budget","Vezes Maior"]][0:10]
```

***Ao atentar-se ao DataFrame acima é pecepitível que Paranormal Activity é o filme com o maior arrecadação em relação ao orçamento de produção,com uma incrível arrecadação 12.918 vezes maior que o custo de produção.***
